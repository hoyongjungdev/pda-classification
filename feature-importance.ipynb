```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix

PREFIX = ''

x = np.load(PREFIX + 'data/x.npy')

tv_x = np.load(PREFIX + 'data/tv_x.npy')
tv_y = np.load(PREFIX + 'data/tv_y.npy')

train_x = tv_x[:,-1,:]
train_y = tv_y.copy()

test_x = np.load(PREFIX + 'data/test_x.npy')
test_y = np.load(PREFIX + 'data/test_y.npy')

test_x = test_x[:,-1,:]

one = (train_y == 1).sum()
zero = (train_y == 0).sum()

ratio = one / zero
print(ratio)
#print(train_x[0])

clf = RandomForestClassifier(random_state=0, n_estimators=1000,class_weight={
    0: 1/ratio
})

clf.fit(train_x, train_y)
predicted_y = clf.predict(test_x)
```

```python
size = len(test_y)
dim = train_x.shape[1]

print((predicted_y == test_y).sum())
print(len(test_y))

print("accuracy: {}".format((predicted_y == test_y).sum() / len(test_y)))
print("f1 score: {}".format(f1_score(test_y, predicted_y)))

print(confusion_matrix(test_y, predicted_y) / size)

fi = []
for i in range(dim):
    fi.append((clf.feature_importances_[i], i))

fi = sorted(fi)

print(fi)

for s in [8, 16, 32, 64]:
    subset = fi[-s:]

    result = [b for a, b in subset]
    result = sorted(result)
    result = np.array(result)
    print(result)

    x_subset = x[:, :, result]
    print(x_subset)
    print(x_subset.shape)

    name = '{}data/x-{}.npy'.format(PREFIX, s)

    np.save(name, x_subset)
```
